# CryoSens Analytics

Generic notebook template for loading, preprocessing, visualizing , and analyzing industrial sensor data.

## 1. Setup
Import the public CryoSens API.

```python
from cryosens.api import *
print("CryoSens generic notebook ready")
```

## 2. Load raw data
Select and load the source file.

```python
df_raw  = load_raw_data()
```

## 3. Preprocessing
Clean and prepare the raw dataset.

```python
df_processed = run_preprocessing_pipeline(df_raw)
```

## 4. Rename columns
Review and rename raw columns before further preprocessing.

```python
rename_columns_ui(df_processed)
```

## 5. Unit conversion

Use this step only when a given analysis requires standardized temperature, pressure, or differential pressure units.

If no conversion is needed, skip Them. 

 To convert temperatures run this cell :

```python
df_processed = convert_temperatures(df_processed)
```

```python
df_processed = convert_pressures(df_processed)
```

To convert pressure diff run this cell

```python
df_processed = convert_pdi(df_processed)
```

## Save and reload the cleaned dataset
Save the cleaned dataset if you want to reuse it later without running the full preprocessing workflow again.

```python
save_dataframe(df_processed, "df_processed.parquet")
```

```python
df_processed = load_dataframe("df_processed.parquet")
```

## 6. Split dataframe by Type 

If the dataset contains many sensors, splitting by sensor type can make visualization and exploration easier.

```python
df_t , df_p , df_pd = split_df_by_sensor_types(df_processed)
```

## 7. Signal exploration
Explore the raw sensor signals before running event detection.

By default, this dashboard is launched on the temperature dataframe (`df_t`) for easier exploration.

These dashboards can run without specifying `unit` or `value_label`.

However, in that case the plots use generic axis labels.

For clearer and more physically meaningful figures, it is recommended to provide:
- `value_label` to describe the measured quantity
- `unit` to display the correct engineering unit on the axis

This is especially useful when working with a homogeneous dataframe such as:
- `df_t` for temperature sensors
- `df_p` for pressure sensors
- `df_pd` for differential pressure / pressure drop sensors

If you use `df_processed` with mixed sensor families, generic labels may be preferable unless the dataframe is filtered first.

```python
plot_sensors_dashboard(df_t , unit="°C", value_label="Temperature ")
```
